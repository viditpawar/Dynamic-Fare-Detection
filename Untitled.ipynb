```python
import os
import psycopg2
import pandas as pd
from dotenv import load_dotenv

load_dotenv()

PG_USER = os.getenv("PG_USER")
PG_PASSWORD = os.getenv("PG_PASSWORD")
PG_HOST = os.getenv("PG_HOST")
PG_PORT = os.getenv("PG_PORT")
PG_DATABASE = os.getenv("PG_DATABASE")
PG_SSLMODE = os.getenv("PG_SSLMODE")

conn = psycopg2.connect(
    dbname=PG_DATABASE,
    user=PG_USER,
    password=PG_PASSWORD,
    host=PG_HOST,
    port=PG_PORT,
    sslmode=PG_SSLMODE
)

print("Connected!")
```

```python
query = """
SELECT
    pickup_datetime,
    dropoff_datetime,
    EXTRACT(EPOCH FROM (dropoff_datetime - pickup_datetime)) / 60.0 AS trip_duration_min,
    trip_distance,
    passenger_count,
    PULocationID,
    DOLocationID,
    RatecodeID,
    payment_type,
    fare_amount,
    tip_amount,
    total_amount
FROM taxi_trips_manhattan
WHERE trip_distance > 0
  AND total_amount > 0
  AND passenger_count > 0
  -- optional: limit data size while experimenting
  -- LIMIT 2_000_000
;
"""

df = pd.read_sql_query(query, conn)
df.head()
len(df)
```

```python
raw_df = df.copy()
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
import holidays

# 1. Start from your original df with total_amount, trip_distance, etc.
df = raw_df.copy()

# now do all your filtering / clipping on df
df = df[
    (df["trip_distance"] > 0) &
    (df["trip_duration_min"] > 0) &
    (df["total_amount"] > 0)
]

fare_hi = df["total_amount"].quantile(0.999)
print("Clipping threshold:", fare_hi)

df = df[df["total_amount"] <= fare_hi].copy()

# 3. Sort by time and rebuild time features
df["pickup_datetime"] = pd.to_datetime(df["pickup_datetime"])
df = df.sort_values("pickup_datetime").reset_index(drop=True)

df["pickup_hour"] = df["pickup_datetime"].dt.hour
df["pickup_dow"]  = df["pickup_datetime"].dt.dayofweek
df["pickup_month"] = df["pickup_datetime"].dt.month
df["pickup_year"]  = df["pickup_datetime"].dt.year
df["pickup_weekofyear"] = df["pickup_datetime"].dt.isocalendar().week.astype(int)

# feature_cols = [
#     "trip_distance",
# #    "trip_duration_min",
#     "passenger_count",
#     "pulocationid",
#     "dolocationid",
# #    "ratecodeid",
# #    "payment_type",
#     "pickup_hour",
#     "pickup_dow",
# ]


feature_cols = [
    "trip_distance",
    "passenger_count",
    "pulocationid",
    "dolocationid",
    "pickup_hour",
    "pickup_dow",
    "pickup_month",
    "pickup_weekofyear",
    "pickup_year",
]
# 4. Log target
df["log_total_amount"] = np.log1p(df["total_amount"])

X = df[feature_cols]
y = df["log_total_amount"]

# 5. Time-based split: first 80% train, last 20% test
split_idx = int(len(df) * 0.8)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# Save original-scale y for metrics
y_train_orig = df["total_amount"].iloc[:split_idx]
y_test_orig  = df["total_amount"].iloc[split_idx:]
```

```python
import xgboost as xgb

xgb_model = xgb.XGBRegressor(
    n_estimators=400,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="reg:squarederror",
    tree_method="hist",   # good default for CPU
    random_state=42,
    n_jobs=-1,
)

xgb_model.fit(X_train, y_train)

# predictions in log space
y_pred_log_xgb  = xgb_model.predict(X_test)

# log-space metrics
rmse_log_xgb = np.sqrt(mean_squared_error(y_test, y_pred_log_xgb))
r2_log_xgb   = r2_score(y_test, y_pred_log_xgb)

# back to original scale
y_pred_orig_xgb = np.expm1(y_pred_log_xgb)

rmse_orig_xgb = np.sqrt(mean_squared_error(y_test_orig, y_pred_orig_xgb))
r2_orig_xgb   = r2_score(y_test_orig, y_pred_orig_xgb)

print("=== XGBoost ===")
print("Log-space  RMSE / RÂ² :", rmse_log_xgb, r2_log_xgb)
print("Original   RMSE / RÂ² :", rmse_orig_xgb, r2_orig_xgb)
```

```python
import lightgbm as lgb

lgbm_model = lgb.LGBMRegressor(
    n_estimators=500,
    learning_rate=0.05,
    num_leaves=64,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1,
)

lgbm_model.fit(X_train, y_train)

y_pred_log_lgb = lgbm_model.predict(X_test)

rmse_log_lgb = np.sqrt(mean_squared_error(y_test, y_pred_log_lgb))
r2_log_lgb   = r2_score(y_test, y_pred_log_lgb)

y_pred_orig_lgb = np.expm1(y_pred_log_lgb)

rmse_orig_lgb = np.sqrt(mean_squared_error(y_test_orig, y_pred_orig_lgb))
r2_orig_lgb   = r2_score(y_test_orig, y_pred_orig_lgb)

print("=== LightGBM ===")
print("Log-space  RMSE / RÂ² :", rmse_log_lgb, r2_log_lgb)
print("Original   RMSE / RÂ² :", rmse_orig_lgb, r2_orig_lgb)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

# Baseline: always predict mean fare from *train* set
baseline_mean = y_train_orig.mean()
baseline_pred = np.full_like(y_test_orig, baseline_mean, dtype=float)

baseline_rmse = np.sqrt(mean_squared_error(y_test_orig, baseline_pred))
baseline_r2   = r2_score(y_test_orig, baseline_pred)

mae_lgb  = mean_absolute_error(y_test_orig, y_pred_orig_lgb)
mape_lgb = np.mean(
    np.abs(y_test_orig - y_pred_orig_lgb) / y_test_orig
) * 100

print("=== Baseline (mean predictor) ===")
print("RMSE:", baseline_rmse)
print("RÂ²  :", baseline_r2)

print("\n=== LightGBM ===")
print("RMSE:", 5.228384376573912)   # or recompute rmse_orig_lgb
print("RÂ²  :", 0.9414625868184862)  # or use r2_orig_lgb
print("MAE :", mae_lgb)
print("MAPE:", mape_lgb, "%")
```
